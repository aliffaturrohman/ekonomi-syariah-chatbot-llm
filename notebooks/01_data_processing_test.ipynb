```python
import pymupdf
import os
def split_pdf_pymupdf(input_file, output_file, start_page, end_page):
    """Memisahkan halaman tertentu dari file PDF."""
    from_page_index = start_page - 1
    to_page_index = end_page - 1
    
    doc = pymupdf.open(input_file)
    new_doc = pymupdf.open()
    new_doc.insert_pdf(doc, from_page=from_page_index, to_page=to_page_index)
    new_doc.save(output_file)
    new_doc.close()
    doc.close()
    
    print(f"File '{output_file}' berhasil dibuat dengan halaman {start_page} s/d {end_page}.")


def batch_split_pdf(input_file, page_ranges):
    base_name = os.path.splitext(os.path.basename(input_file))[0]
    for title, (start, end) in page_ranges.items():
        safe_title = title.replace(" ", "_").replace("/", "-")
        output_path = f"../data/raw/pdf_processed/{base_name}_{safe_title}_[{start}-{end}].pdf"
        split_pdf_pymupdf(input_file, output_path, start, end)

input_path = "../data/raw/pdf/Ekonomi Syariah E-Book [UINSA].pdf"

page_ranges = {
    "Bab 1": [12, 26],
    "Bab 2": [30, 38],
    "Bab 3": [42, 54],
    "Bab 4": [58, 73],
    "Bab 5": [78, 94],
    "Bab 6": [98, 106],
    "Bab 7": [110, 125],
    "Bab 8": [128, 145],
    "Bab 9": [150, 160],
    "Bab 10": [164, 179],
    "Bab 11": [184, 196],
    "Bab 12": [206, 211],
}

batch_split_pdf(input_path, page_ranges)
```

## Marker pipeline

```python
import os
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered

file_path = "../data/raw/pdf_processed/Ekonomi Syariah E-Book [UINSA]_Bab_2_[30-38].pdf"
file_name = os.path.splitext(os.path.basename(file_path))[0]
print(f"Nama file: {file_name}")

converter = PdfConverter(artifact_dict=create_model_dict())
rendered = converter(file_path)
text, _, images = text_from_rendered(rendered)

# Pastikan direktori output ada
output_dir = "../data/raw/pdf_finalize"
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, f"{file_name}.md")

with open(output_path, "w", encoding="utf-8") as f:
    f.write(text)

print(f"File disimpan ke: {output_path}")
```

## Docsling Pipeline

```python
from docling.datamodel.base_models import InputFormat
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.pipeline.vlm_pipeline import VlmPipeline
from docling.datamodel.pipeline_options import VlmPipelineOptions
from docling.datamodel import vlm_model_specs

pipeline_options = VlmPipelineOptions(
    vlm_options=vlm_model_specs.GRANITEDOCLING_TRANSFORMERS,  
)

converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(
            pipeline_cls=VlmPipeline,
            pipeline_options=pipeline_options,
        ),
    }
)

doc = converter.convert(source="../data/raw/pdf_processed/Ekonomi Syariah E-Book [UINSA]_Bab_2_[30-38].pdf").document
print(doc.export_to_markdown())
```
